```python
import pandas as pd
df_sales_records = pd.read_csv('/content/1000 Sales Records.csv')

df_sales_records
```

```python
print(df_sales_records.info())
print(df_sales_records.describe())
```

```python
print(df_sales_records.isnull().sum())
```

```python
#Sales by Region
sales_by_region = df_sales_records.groupby('Region')['Total Revenue'].sum()

print(sales_by_region)
```

```python
#Total Revenue by Country
total_revenue_by_country = df_sales_records.groupby('Country')['Total Revenue'].sum()

print(total_revenue_by_country)
```

```python
#Average Profit by Item Type
average_profit_by_item_type = df_sales_records.groupby('Item Type')['Total Profit'].mean()

print(average_profit_by_item_type)
```

```python
#Correlation Analysis
correlation_matrix = df_sales_records.select_dtypes(include=['number']).corr()

print(correlation_matrix)
```

```python
#Top 10 Orders by Total Revenue
top_10_orders = df_sales_records.nlargest(10, 'Total Revenue')

print(top_10_orders)
```

```python
top_orders = df_sales_records.sort_values(by='Total Revenue', ascending=False).head(10)

print(top_orders)
```

```python
#Orders with High Priority
high_priority_orders = df_sales_records[df_sales_records['Order Priority'] == 'H']

print(high_priority_orders)
```

```python
#Average Sales
average_sales = df_sales_records['Total Revenue'].mean()

print(average_sales)
```

```python
#number of orders per country
orders_per_country = df_sales_records.groupby('Country')['Order ID'].count()

print(orders_per_country)
```

```python
orders_by_country = df_sales_records['Country'].value_counts()
print(orders_by_country)
```

```python
#Filter data for a specific region
region_data_filter = df_sales_records[df_sales_records['Region'] == 'North America']

print(region_data_filter)
```

```python
#least common item type
least_common_item_type = df_sales_records['Item Type'].value_counts().idxmin()

print(least_common_item_type)
```

```python
item_counts = df_sales_records['Item Type'].value_counts()
least_common_item = item_counts.idxmin()

print(f"The least common item type is: {least_common_item}")
```

```python
#Most common item type
most_common_item_type = df_sales_records['Item Type'].value_counts().idxmax()

print(most_common_item_type)
```

```python
most_common_item = item_counts.idxmax()

print(f"The most common item type is: {most_common_item}")
```
